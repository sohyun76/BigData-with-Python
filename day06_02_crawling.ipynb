```python
import requests as req
from bs4 import BeautifulSoup
```

```python
# 요청하는 사이트에 컴퓨터(크롤링)이 아니라 사람이 요청하는 것이라고 속이기
# https://www.whatismybrowser.com/detect/what-is-my-user-agent 파란색네모 복사하기
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
```

```python
url = 'http://www.coupang.com/np/search?component=&q=%EB%A7%88%EC%8A%A4%ED%81%AC&channel=user'
url
```

```python
response = req.get(url, headers=headers)
```

```python
parseHtml = BeautifulSoup(response.text, 'html.parser') 
parseHtml
```

```python
product_list = parseHtml.select('#productList')[0]
item_li = product_list.findAll('li')
```

```python
for item in item_li:
    item_title = item.select('.name')[0].text
    item_price = item.select('.price-value')[0].text
    print('{0} / 가격: {1}원'.format(item_title, item_price))
# ex) 일회용 마스크 블랙 50매 / 가격 : 35000
```
