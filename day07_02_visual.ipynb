## day07_02_visual
- 데이터 시각화 예제

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
# 시각화 함수 호출 시에 바로 notebook에 표현하기
%matplotlib inline 
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sepal_length[:20].head()
```

### bar: 데이터의 갯수를 셀 때 유용한 plot

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title('sepal_length visual')
plt.ylabel('data')
plt.xlabel('sepal_length')
```

```python
# 붓꽃 종의 데이터의 평균치를 이용하여 bar plot 그리기
iris.groupby(iris.species).mean().plot(kind='bar')
```

## pie plot: 전체 중에서 비율을 알고 싶은 경우에 유용

```python
# 붓꽃 데이터의 종의 갯수 구하기
df3 = iris.species.value_counts()
```

```python
df3
```

```python
df3.plot.pie(autopct='%.2f%%')
```

```python
df4 = iris.reindex(np.random.permutation(iris.index)[0:40])
```

```python
df4.species.value_counts()
```

```python
df5 = df4.species.value_counts()
df5.plot.pie(autopct='%.2f%%')
```

## hist(histogram): 빈도 수

```python
iris.plot.hist(rwidth=0.9) # rwidth bar의 넓이
```

# tips 데이터 분석하기

```python
tips = sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
# 성별에 따른 팁의 금액
sns.relplot(x='sex', y='tip', data=tips)
```

```python
# 남녀의 비율에 따른 팁
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# 흡연여부에 따른 팁 금액
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill'
            , y='tip'
            , data=tips
            , hue='smoker'
            , size='size'
           )
```

```python
# 점심과 저녁으로 나누기
# - 카테고리로 나누기
sns.catplot(x='time', y='total_bill', data=tips, hue='smoker')
```
