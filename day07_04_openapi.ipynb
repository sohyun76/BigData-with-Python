## day07_04_openapi
- 영화진흥원 API를 이용하여 분석 데이터 수집하기

```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
# 데이터를 모을 api 의 주소
target_date = '20200319'
url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=430156241533f1d058c603178cc3ca0e&targetDt=' + target_date
```

```python
# url로 지정된 곳으로 부터 데이터 수신 받기
res = requests.get(url)
```

```python
data = json.loads(res.text)
data
```

### 수집할 데이터
- rank(순위)
- rankOldAndNew(신규진입여부)
- movieCd(영화코드)
- movieNm(영화이름)
- salesAmt(매출액)
- audiCnt(관객수)
- scrnCnt(상영스크린수)
- showCnt(상영횟수)

```python
items = data['boxOfficeResult']['dailyBoxOfficeList']
```

```python
key_list = ['rank', 'movieNm', 
            'audiCnt', 'rankOldAndNew', 
            'movieCd', 'salesAmt', 
            'scrnCnt', 'showCnt'] 
```

```python
# 실습하기 for문을 이용하여 items의 데이터를 보여주세요.
# - movieNm만 보여주세요.
movie_list = []  # 수신받은 데이터를 데이터프레임 형태로 변경하기 위해서 값을 모으는 리스트

for item in items:
    tmp_list = []
    for key in key_list:  
        tmp_list.append(item[key])
    tmp_list.append(target_date)
    movie_list.append(tmp_list)
key_list.append('targetDate')        
```

```python
pd.DataFrame(movie_list, columns=key_list)
```
