```python
# 크롤링 예제 - 쿠차, 옐로우모바일, 구글, 에어비앤비
# https://www.apache.org/ 사이트의 카테고리 정보 수집하기
# 외부 패키지 설치하기
# id(#), class(.)
```

```python
# 크롤링을 위한 외부 패키지 설치
# -python에서 웹 통신을 위한 패키지 사용하기
import requests as req
from bs4 import BeautifulSoup
```

```python
# 1. 긁어오고 싶은 페이지 지정
url = 'https://www.apache.org/'
# url 문자열로 넣어주기.
```

```python
# 2. 페이지 요청하기
# - 요청한다는 것은 해당 페이지에 보여달라고 요구하는 것이다.
# req.get(url) -> <Response [200]> 출력
parseHtml = req.get(url).text
```

```python
# - 404 : 페이지 없음 코드
# - 200 : 성공했다는 의미의 코드
```

```python
# 3. 요청된 페이지로부터 html 코드화시키기
soup = BeautifulSoup(parseHtml, 'html.parser')
```

```python
# 4. BY NAME의 정보만 soup 변수에서 조회하기.
# soup.select('#by_name > .col-lg-4 >.row >.col-lg-6 >.list-unstyled' )
cate = soup.select('.list-unstyled > li > a' )
```

```python
# 5. cate를 반복문을 통하여 카테고리 리스트 만들기
# len(cate) -> 268
for item in cate :
    print(item)
```
