```python
#긁어올 페이지 지정
my_keyword = '머신러닝'
url = 'http://jtbc.joins.com/search/news?term=' + my_keyword
print(url)
```

```python
# 긁어올 페이지로부터 파이썬을 이용하여 데이터 긁어오기
response = req.get(url)
response
```

```python
# response 객체에서 text보기
parseHtml = BeautifulSoup(response.text, 'html.parser')
parseHtml
```

```python
#크롤링하고 시픈 데이터 저장하기
# - cate = parseHtml.select('.wrap_sch_area>.txt_result>strong')
parseHtml.select('.txt_result')[0].find('strong').text+'건'
```
